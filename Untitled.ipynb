```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
cotiz = [1,3.69,3,3.05,2.94,3.15,3.26, 3.08, 3.9, 4, 4.27, 4.68, 5.71, 14.10, 15.45, 15.55, 18.97, 41]
```

```python
rel_cotiz = []
for i,c in enumerate(cotiz):
    if 0 <= i < 2:
        rel_cotiz.append(c)
    elif 2 <= i < 6:
        rel_cotiz.append(cotiz[i]/cotiz[2])
    elif 6 <= i < 14:
        rel_cotiz.append(cotiz[i]/cotiz[6])
    elif 14 <= i:
        rel_cotiz.append(cotiz[i]/cotiz[14])

```

```python
plt.plot(range(2001,2003),rel_cotiz[0:2], label='Pre-K', c='red')
plt.plot(range(2003,2007),rel_cotiz[2:6], label='NK', c='blue')
plt.plot(range(2007,2015),rel_cotiz[6:14], label='CFK', c='skyblue')
plt.plot(range(2015,2019),rel_cotiz[14:], label='MM', c='y')
plt.legend()
plt.xlabel('AÃ±o')
plt.ylabel('Crecimiento rel. USD/ARS desde toma del poder')
plt.xticks(range(2001,2019), rotation=45)
plt.show()
```
